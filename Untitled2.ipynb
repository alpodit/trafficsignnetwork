```python
import albumentations as A
from albumentations.pytorch import ToTensorV2

IMG_SIZE = 32  # or 64 if you prefer

train_transforms = A.Compose([
    # 1) Resize every image to IMG_SIZE×IMG_SIZE
    A.Resize(IMG_SIZE, IMG_SIZE),

    # 2) Geometric + photometric augmentations
    A.Affine(
        translate_percent=0.1,
        scale=(0.8, 1.2),
        rotate=(-15, 15),
        p=0.7
    ),
    A.RandomBrightnessContrast(
        brightness_limit=0.3,
        contrast_limit=0.3,
        p=0.7
    ),
    A.GaussNoise(
        std_range=(0.04, 0.2),
        mean_range=(0.0, 0.0),
        per_channel=True,
        noise_scale_factor=1.0,
        p=0.5
    ),
    A.MotionBlur(blur_limit=3, p=0.3),
    A.OneOf([
        A.RandomFog(p=0.2),
        A.RandomRain(p=0.2),
        A.RandomSnow(p=0.2),
    ], p=0.3),

    # 3) Normalize & to-tensor
    A.Normalize(mean=(0.485, 0.456, 0.406),
                std=(0.229, 0.224, 0.225)),
    ToTensorV2(),
])

val_transforms = A.Compose([
    A.Resize(IMG_SIZE, IMG_SIZE),
    A.Normalize(mean=(0.485, 0.456, 0.406),
                std=(0.229, 0.224, 0.225)),
    ToTensorV2(),
])
```

```python
import os
from torch.utils.data import Dataset

class GTSRBDataset(Dataset):
    def __init__(self, root_dir, transforms=None):
        """
        root_dir: path to folder containing class subfolders '00000', '00001', ... '00042'
        transforms: Albumentations pipeline
        """
        self.transforms = transforms
        self.images = []
        self.labels = []

        # Walk the class folders directly under root_dir
        for class_id in sorted(os.listdir(root_dir)):
            class_dir = os.path.join(root_dir, class_id)
            if not os.path.isdir(class_dir):
                continue
            # Gather every .ppm image in that class directory
            for img_name in os.listdir(class_dir):
                if img_name.lower().endswith('.ppm'):
                    self.images.append(os.path.join(class_dir, img_name))
                    self.labels.append(int(class_id))

        assert len(self.images) == len(self.labels), "Images/Labels mismatch"

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        import cv2
        img_path, label = self.images[idx], self.labels[idx]
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        if self.transforms:
            augmented = self.transforms(image=img)
            img = augmented['image']

        return img, label
```

```python
import numpy as np
from torch.utils.data import Subset, DataLoader


# Point to the folder that contains 00000–00042
train_base = GTSRBDataset(
    root_dir='/content/drive/MyDrive/GTSRB/GTSRB/Final_Training/Images',
    transforms=train_transforms
)
val_base = GTSRBDataset(
    root_dir='/content/drive/MyDrive/GTSRB/GTSRB/Final_Training/Images',
    transforms=val_transforms
)


from torch.utils.data import Subset, DataLoader

# 2) Shuffle+split indices for 80/20 train/val
num_samples = len(train_base)
indices     = np.arange(num_samples)
np.random.seed(42)
np.random.shuffle(indices)

split = int(0.8 * num_samples)
train_idx, val_idx = indices[:split], indices[split:]

# 3) Wrap into Subsets (each with its own transforms pipeline)
train_ds = Subset(train_base, train_idx)
val_ds   = Subset(val_base,   val_idx)

# 4) Create loaders
train_loader = DataLoader(train_ds, batch_size=64, shuffle=True,  num_workers=2, pin_memory=True)
val_loader   = DataLoader(val_ds,   batch_size=64, shuffle=False, num_workers=2, pin_memory=True)

```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

# ── Basic residual block ───────────────────────────────────────────────────────
class BasicBlock(nn.Module):
    expansion = 1

    def __init__(self, in_planes, planes, stride=1):
        super(BasicBlock, self).__init__()
        self.conv1 = nn.Conv2d(in_planes, planes,
                               kernel_size=3, stride=stride,
                               padding=1, bias=False)
        self.bn1 = nn.BatchNorm2d(planes)
        self.conv2 = nn.Conv2d(planes, planes,
                               kernel_size=3, stride=1,
                               padding=1, bias=False)
        self.bn2 = nn.BatchNorm2d(planes)

        # if dimensions change, use a 1×1 conv for the skip path
        self.shortcut = nn.Sequential()
        if stride != 1 or in_planes != planes * self.expansion:
            self.shortcut = nn.Sequential(
                nn.Conv2d(in_planes,
                          planes * self.expansion,
                          kernel_size=1,
                          stride=stride,
                          bias=False),
                nn.BatchNorm2d(planes * self.expansion)
            )

    def forward(self, x):
        out = F.relu(self.bn1(self.conv1(x)))
        out = self.bn2(self.conv2(out))
        out += self.shortcut(x)
        return F.relu(out)


# ── ResNet with four stages of BasicBlock ──────────────────────────────────────
class ResNet(nn.Module):
    def __init__(self, block, num_blocks, num_classes=43):
        super(ResNet, self).__init__()
        self.in_planes = 64

        # initial conv + BN
        self.conv1 = nn.Conv2d(3, 64,
                               kernel_size=3,
                               stride=1,
                               padding=1,
                               bias=False)
        self.bn1 = nn.BatchNorm2d(64)

        # four stages: [2,2,2,2] blocks by default for ResNet-18
        self.layer1 = self._make_layer(block,  64, num_blocks[0], stride=1)
        self.layer2 = self._make_layer(block, 128, num_blocks[1], stride=2)
        self.layer3 = self._make_layer(block, 256, num_blocks[2], stride=2)
        self.layer4 = self._make_layer(block, 512, num_blocks[3], stride=2)

        self.avgpool = nn.AdaptiveAvgPool2d((1, 1))
        self.dropout = nn.Dropout(0.5)
        self.fc = nn.Linear(512 * block.expansion, num_classes)

    def _make_layer(self, block, planes, blocks, stride):
        strides = [stride] + [1] * (blocks - 1)
        layers = []
        for s in strides:
            layers.append(block(self.in_planes, planes, s))
            self.in_planes = planes * block.expansion
        return nn.Sequential(*layers)

    def forward(self, x):
        out = F.relu(self.bn1(self.conv1(x)))
        out = self.layer1(out)
        out = self.layer2(out)
        out = self.layer3(out)
        out = self.layer4(out)
        out = self.avgpool(out)
        out = out.view(out.size(0), -1)
        out = self.dropout(out)
        return self.fc(out)


# ── Helper to instantiate ResNet-18 ────────────────────────────────────────────
def resnet18(num_classes=43):
    return ResNet(BasicBlock, [2, 2, 2, 2], num_classes)


# ── Move to device ─────────────────────────────────────────────────────────────
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = resnet18().to(device)
print(model)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.amp import GradScaler
from torch.cuda.amp import autocast
from torch.optim.lr_scheduler import OneCycleLR
from tqdm import tqdm

# 1) Loss, optimizer, scheduler, AMP scaler
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(
    model.parameters(),
    lr=1e-3,
    momentum=0.9,
    weight_decay=1e-4
)
scheduler = OneCycleLR(
    optimizer,
    max_lr=1e-2,
    steps_per_epoch=len(train_loader),
    epochs=20,
    pct_start=0.3,
    anneal_strategy='cos'
)
scaler = GradScaler()

# 2) One epoch training with AMP + scheduler.step() per batch
def train_one_epoch(model, loader, criterion, optimizer, scheduler, scaler, device):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in tqdm(loader, desc="Train"):
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad()

        with autocast():
            outputs = model(images)
            loss = criterion(outputs, labels)

        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()
        scheduler.step()

        running_loss += loss.item() * images.size(0)
        preds = outputs.argmax(dim=1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

    return running_loss / total, correct / total

# 3) Validation (no AMP, no scheduler)
def validate(model, loader, criterion, device):
    model.eval()
    running_loss = 0.0
    correct = 0
    total = 0

    with torch.no_grad():
        for images, labels in tqdm(loader, desc="Validate"):
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)

            running_loss += loss.item() * images.size(0)
            preds = outputs.argmax(dim=1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)

    return running_loss / total, correct / total

# 4) Training loop with early stopping & checkpointing
best_val_acc = 0.0
patience = 5
epochs_no_improve = 0
num_epochs = 20

for epoch in range(1, num_epochs + 1):
    train_loss, train_acc = train_one_epoch(
        model, train_loader, criterion,
        optimizer, scheduler, scaler, device
    )
    val_loss, val_acc = validate(
        model, val_loader, criterion, device
    )

    print(f"Epoch {epoch:02d} | "
          f"Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f} | "
          f"Val Loss:   {val_loss:.4f},   Val Acc: {val_acc:.4f}")

    # Check for improvement
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        epochs_no_improve = 0
        torch.save(model.state_dict(), '/content/drive/MyDrive/best_resnet18.pth')
        print(f"→ New best model saved (Val Acc: {best_val_acc:.4f})")
    else:
        epochs_no_improve += 1
        print(f"→ No improvement for {epochs_no_improve}/{patience} epochs")
        if epochs_no_improve >= patience:
            print(f"→ Early stopping at epoch {epoch}")
            break

print(f"Training complete. Best Val Acc: {best_val_acc:.4f}")
```

```python
import os
import csv
import cv2
import torch
from torch.utils.data import Dataset, DataLoader
from tqdm import tqdm

# ── 1) Dataset for the GTSRB Test Set ───────────────────────────────────────────
class GTSRBTestDataset(Dataset):
    def __init__(self, images_dir, gt_csv, transforms=None):
        """
        images_dir: path to folder containing all .ppm test images
        gt_csv:     path to GT-final_test.csv with semicolon delimiters
        transforms: Albumentations pipeline (resize + normalize + ToTensorV2)
        """
        self.images_dir = images_dir
        self.transforms = transforms
        self.images = []
        self.labels = []

        # Read the CSV
        with open(gt_csv, newline='') as f:
            reader = csv.reader(f, delimiter=';')
            header = next(reader)  # skip header
            for row in reader:
                filename = row[0]            # e.g. '12629.ppm'
                class_id = int(row[-1])      # last column is ClassId
                path = os.path.join(images_dir, filename)
                # Only add if file actually exists
                if os.path.isfile(path):
                    self.images.append(path)
                    self.labels.append(class_id)

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        img_path = self.images[idx]
        label = self.labels[idx]
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        if self.transforms:
            img = self.transforms(image=img)['image']
        return img, label


# ── 2) Instantiate Test Dataset & Loader ───────────────────────────────────────
# Adjust these paths if your files live elsewhere:
test_images_dir = '/content/drive/MyDrive/GTSRB/GTSRB/Final_Test/Images'
gt_csv_path     = '/content/drive/MyDrive/GTSRB/GT-final_test.csv'

test_ds = GTSRBTestDataset(
    images_dir=test_images_dir,
    gt_csv=gt_csv_path,
    transforms=val_transforms
)
test_loader = DataLoader(
    test_ds,
    batch_size=64,
    shuffle=False,
    num_workers=2,
    pin_memory=True
)

# Sanity check: should match number of entries in CSV
print(f"Number of test samples: {len(test_ds)}")


# ── 3) Load Best Model & Evaluate ───────────────────────────────────────────────
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.load_state_dict(torch.load('/content/drive/MyDrive/best_resnet18.pth', map_location=device))
model.to(device).eval()

correct = 0
total   = 0

with torch.no_grad():
    for imgs, labels in tqdm(test_loader, desc="Test"):
        imgs, labels = imgs.to(device), labels.to(device)
        preds = model(imgs).argmax(dim=1)
        correct += (preds == labels).sum().item()
        total   += labels.size(0)

print(f"Test Accuracy: {correct/total:.4%}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import torch

# 1) Make sure your model and device are set
#    (e.g. model, device = your trained model and torch.device)
model = model.to(device)
model.eval()

# 2) Mapping from class index to human-readable label
CLASS_NAMES = {
    0:  "Speed limit (20km/h)",      1:  "Speed limit (30km/h)",
    2:  "Speed limit (50km/h)",      3:  "Speed limit (60km/h)",
    4:  "Speed limit (70km/h)",      5:  "Speed limit (80km/h)",
    6:  "End of speed limit (80km/h)",7:  "Speed limit (100km/h)",
    8:  "Speed limit (120km/h)",     9:  "No passing",
    10: "No passing: >3.5t",         11: "Right-of-way next intersection",
    12: "Priority road",             13: "Yield",
    14: "Stop",                      15: "No vehicles",
    16: "Vehicles >3.5t prohibited", 17: "No entry",
    18: "General caution",           19: "Danger curve left",
    20: "Danger curve right",        21: "Double curve",
    22: "Bumpy road",                23: "Slippery road",
    24: "Road narrows right",        25: "Road work",
    26: "Traffic signals",           27: "Pedestrian crossing",
    28: "School crossing",           29: "Bicycles crossing",
    30: "Snow/Ice",                  31: "Wild animals",
    32: "End all limits",            33: "Turn right ahead",
    34: "Turn left ahead",           35: "Ahead only",
    36: "Straight or right",         37: "Straight or left",
    38: "Keep right",                39: "Keep left",
    40: "Roundabout",                41: "End no passing",
    42: "End no passing: >3.5t"
}

# 3) De-normalization params (must match your val_transforms)
mean = np.array([0.485, 0.456, 0.406])
std  = np.array([0.229, 0.224, 0.225])

# 4) The helper function
def show_predictions(dataset, model, device, indices):
    """
    dataset: your test Dataset (returns (tensor, label))
    model:   trained ResNet on device
    device:  torch.device('cuda') or ('cpu')
    indices: list of integer indices to display
    """
    model.to(device)
    model.eval()
    n = len(indices)
    cols = min(4, n)
    rows = (n + cols - 1) // cols
    fig, axes = plt.subplots(rows, cols, figsize=(4*cols, 3*rows))
    axes = axes.flatten()

    with torch.no_grad():
        for ax, idx in zip(axes, indices):
            img_tensor, true_label = dataset[idx]
            # De-normalize & HWC
            img_np = img_tensor.cpu().numpy().transpose(1,2,0) * std + mean
            img_np = np.clip(img_np, 0, 1)

            # Predict
            inp = img_tensor.unsqueeze(0).to(device)
            out = model(inp)
            pred_label = out.argmax(dim=1).item()

            ax.imshow(img_np)
            ax.axis('off')
            ax.set_title(f"T: {CLASS_NAMES[true_label]}\nP: {CLASS_NAMES[pred_label]}")

    # hide extra axes
    for ax in axes[n:]:
        ax.axis('off')
    plt.tight_layout()
    plt.show()

# 5) Example: show first 8 test images
show_predictions(test_ds, model, device, list(range(8)))
```

```python
import cv2
import torch
import numpy as np

# Load the TorchScript model
model_ts = torch.jit.load('/content/drive/MyDrive/resnet18_ts.pt').to(device).eval()

# Precompute mean/std for normalization
mean = np.array([0.485, 0.456, 0.406])
std  = np.array([0.229, 0.224, 0.225])


cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise RuntimeError("Could not open camera.")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Preprocess
    img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (32, 32))
    img = img.astype(np.float32) / 255.0
    img = (img - mean) / std
    tensor = torch.from_numpy(img).permute(2,0,1).unsqueeze(0).to(device)

    # Inference
    with torch.no_grad():
        out = model_ts(tensor)
        pred = out.argmax(dim=1).item()

    # Overlay
    label = CLASS_NAMES[pred]
    cv2.putText(frame, label, (10,30),
                cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0,255,0), 2)
    cv2.imshow("Traffic Sign Demo", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
